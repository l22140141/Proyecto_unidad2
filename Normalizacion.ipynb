#Normalización Paola Martínez

```python
import pandas as pd
```

```python
URL_MSEASONS      = "https://raw.githubusercontent.com/l22140141/Proyecto_unidad2/refs/heads/main/BASES%20DE%20DATOS/MSeasons.csv"
URL_WCONF_TOURNEY= "https://raw.githubusercontent.com/l22140141/Proyecto_unidad2/refs/heads/main/BASES%20DE%20DATOS/WConferenceTourneyGames.csv"
mseasons_raw      = pd.read_csv(URL_MSEASONS)
wconf_tourney_raw = pd.read_csv(URL_WCONF_TOURNEY)
print("Archivos cargados desde GitHub")
print(f"  MSeasons:               {mseasons_raw.shape[0]} filas × {mseasons_raw.shape[1]} cols")
print(f"  WConferenceTourneyGames:{wconf_tourney_raw.shape[0]} filas × {wconf_tourney_raw.shape[1]} cols")
```

```python
# PASO 2: Desnormalizar
df_desnorm = wconf_tourney_raw.merge(mseasons_raw, on="Season", how="left")

# PASO 3: Normalizar — MSeasons
cols_seasons  = list(mseasons_raw.columns)
mseasons_norm = df_desnorm[cols_seasons].drop_duplicates().reset_index(drop=True)

# PASO 4: Normalizar — WConferenceTourneyGames
cols_tourney = list(wconf_tourney_raw.columns)
wconf_norm   = df_desnorm[cols_tourney].drop_duplicates().reset_index(drop=True)
wconf_norm.insert(0, "game_id", wconf_norm.index + 1)

# PASO 5: Verificar integridad
df_verificacion  = wconf_norm.drop(columns=["game_id"]).merge(mseasons_norm, on="Season", how="left")
columnas_comunes = [c for c in df_desnorm.columns if c in df_verificacion.columns]
es_igual = (
    df_desnorm[columnas_comunes].reset_index(drop=True)
    .equals(df_verificacion[columnas_comunes].reset_index(drop=True))
)

# PASO 6: Resultados
print("\nTABLA NORMALIZADA 1 — MSeasons")
print(mseasons_norm)

print("\nTABLA NORMALIZADA 2 — WConferenceTourneyGames")
print(wconf_norm.head(10))

print(f"\n¿Integridad verificada? → {'✔ SÍ' if es_igual else '✘ NO'}")

celdas_antes   = df_desnorm.shape[0] * df_desnorm.shape[1]
celdas_despues = mseasons_norm.shape[0] * mseasons_norm.shape[1] + wconf_norm.shape[0] * wconf_norm.shape[1]
print(f"Reducción de redundancia: {((celdas_antes - celdas_despues) / celdas_antes * 100):.1f}%")

# PASO 7: Guardar en Excel
with pd.ExcelWriter("resultado_normalizacion.xlsx", engine="openpyxl") as writer:
    mseasons_norm.to_excel(writer, sheet_name="MSeasons", index=False)
    wconf_norm.to_excel(writer, sheet_name="WConferenceTourney", index=False)
    df_desnorm.to_excel(writer, sheet_name="Desnormalizada", index=False)

print("\nArchivo Excel guardado como: resultado_normalizacion.xlsx")
```

#Desnormalización Paola Martínez

```python
import pandas as pd
```

```python
URL_MSEASONS      = "https://raw.githubusercontent.com/l22140141/Proyecto_unidad2/refs/heads/main/BASES%20DE%20DATOS/MSeasons.csv"
URL_WCONF_TOURNEY= "https://raw.githubusercontent.com/l22140141/Proyecto_unidad2/refs/heads/main/BASES%20DE%20DATOS/WConferenceTourneyGames.csv"
mseasons_raw      = pd.read_csv(URL_MSEASONS)
wconf_tourney_raw = pd.read_csv(URL_WCONF_TOURNEY)

print("Archivos cargados desde GitHub")
print(f"\nMSeasons: {mseasons_raw.shape[0]} filas × {mseasons_raw.shape[1]} cols")
print(mseasons_raw.head(3))
print(f"\nWConferenceTourneyGames: {wconf_tourney_raw.shape[0]} filas × {wconf_tourney_raw.shape[1]} cols")
print(wconf_tourney_raw.head(3))
```

```python
df_desnorm = wconf_tourney_raw.merge(mseasons_raw, on="Season", how="left")

print("\n" + "=" * 60)
print("TABLA DESNORMALIZADA (WConferenceTourneyGames + MSeasons)")
print("=" * 60)
print(f"Dimensiones: {df_desnorm.shape[0]} filas × {df_desnorm.shape[1]} columnas")
print(df_desnorm.head(10))
```

```python
cols_seasons = [c for c in mseasons_raw.columns if c != "Season"]
print("\n COLUMNAS REPETIDAS DE MSeasons EN LA TABLA DESNORMALIZADA:")
for col in cols_seasons:
    repeticiones = df_desnorm[col].notna().sum()
    unicos       = df_desnorm["Season"].nunique()
    print(f"  '{col}' aparece {repeticiones} veces — solo {unicos} valores únicos")

with pd.ExcelWriter("desnormalizacion.xlsx", engine="openpyxl") as writer:
    mseasons_raw.to_excel(writer, sheet_name="MSeasons_original", index=False)
    wconf_tourney_raw.to_excel(writer, sheet_name="WConference_original", index=False)
    df_desnorm.to_excel(writer, sheet_name="Desnormalizada", index=False)

print("\nArchivo Excel guardado como: desnormalizacion.xlsx")
```

#**NORMALIZACIÓN EVELYN MORANTES**

```python
import pandas as pd

# 1. Tabla 'WGameCities' (Datos de tus juegos)
# Aquí ponemos el listado de lo que contiene tu CSV (ejemplo de las primeras filas)
w_game_data = {
    'Season': [2010, 2010, 2010, 2010, 2010],
    'DayNum': [11, 11, 11, 11, 11],
    'WTeamID': [3103, 3104, 3110, 3111, 3119],
    'LTeamID': [3237, 3399, 3224, 3267, 3447],
    'CRType': ['Regular', 'Regular', 'Regular', 'Regular', 'Regular'],
    'CityID': [4002, 4085, 4363, 4158, 4367]
}
df_w_games = pd.DataFrame(w_game_data)

# 2. Tabla 'MSeasons' (Tus temporadas y regiones)
# Listado basado en el contenido de tu archivo MSeasons
seasons_data = {
    'Season': [1985, 1986, 1987, 1988],
    'DayZero': ['10/29/1984', '10/28/1985', '10/27/1986', '11/02/1987'],
    'RegionW': ['East', 'East', 'East', 'East'],
    'RegionX': ['West', 'Midwest', 'Southeast', 'Midwest'],
    'RegionY': ['Midwest', 'Southeast', 'Midwest', 'Southeast'],
    'RegionZ': ['Southeast', 'West', 'West', 'West']
}
df_seasons_raw = pd.DataFrame(seasons_data)

# --- PROCESO DE NORMALIZACIÓN ---

# A. Normalizar Ciudades (Extraer catálogo de CityID de WGameCities)
# Como tu archivo no tiene nombres de ciudades, normalizamos los IDs únicos
cat_ciudades = df_w_games[['CityID']].drop_duplicates().reset_index(drop=True)

# B. Crear tabla de Hechos (WGameCities normalizada)
w_game_cities = df_w_games.copy()

# C. Normalizar Regiones (Catálogo extraído de tus columnas de MSeasons)
regiones_lista = pd.unique(df_seasons_raw[['RegionW', 'RegionX', 'RegionY', 'RegionZ']].values.ravel())
cat_regiones = pd.DataFrame({'RegionID': range(1, len(regiones_lista) + 1), 'RegionName': regiones_lista})

print("TABLA WGameCities (Normalizada)")
print(w_game_cities)

print("\n TABLA DIM_Ciudades (Catálogo de IDs)")
print(cat_ciudades)

print("\n TABLA DIM_Regiones (Catálogo extraído de tus Seasons)")
print(cat_regiones)
```
