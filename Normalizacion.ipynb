Normalización Samantha V

```python
import pandas as pd
```

Cargar archivos

```python
df_season = pd.read_csv('MSeasons.csv')
df_results = pd.read_csv('WNCAATourneyDetailedResults.csv')
```

```python
print(df_season.head())
print(df_results.head())
```

```python
tabla_seasons = df_season[['Season','DayZero','RegionW','RegionX','RegionY','RegionZ' ]]
print(tabla_seasons.head())
```

```python
tabla_resultados = df_results.copy()
print(tabla_resultados.head())
```

Desnormalización

```python
df_modelo =pd.merge(
    df_results,
    df_season,
    on='Season',
    how='left'
)
print(df_modelo.head())
```

```python
print(df_modelo.columns)
```

```python
df_modelo['DiffScore'] = df_modelo['WScore'] - df_modelo['LScore']
df_modelo['Win'] = 1 #equipo ganador
```

```python
df_modelo.to_csv('dataset_modelo.csv', index=False)
```
